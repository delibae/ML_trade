```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import matplotlib.pyplot as plt
from google.colab import auth
from google.colab import drive
import pandas as pd

filename1 = '/content/drive/MyDrive/BIG DATA/TFN.csv'
filename2 = '/content/drive/MyDrive/BIG DATA/dataN.csv'

TF1 = pd.read_csv(filename1)
data1 = pd.read_csv(filename2)
data2 = data1.loc[150000:199999, :]
data = pd.DataFrame.reset_index(data2)
TF = TF1.loc[150000:199999,:]
gpu = 'cuda' if torch.cuda.is_available() else 'cpu'

# for reproducibility
torch.manual_seed(777)
if gpu == 'cuda':
    torch.cuda.manual_seed_all(777)

x6 = torch.FloatTensor(data['change'].values)
x_6 = x6.view(50000,1)
x7 = torch.FloatTensor(data['change_2'].values)
x_7 = x7.view(50000,1)
x8 = torch.FloatTensor(data['change_3'].values)
x_8 = x8.view(50000,1)
x9 = torch.FloatTensor(data['change_5'].values)
x_9 = x9.view(50000,1)
x1 = torch.FloatTensor(data['change_10'].values)
x_1 = x1.view(50000,1)
x2 = torch.FloatTensor(data['change_20'].values)
x_2 = x1.view(50000,1)
x3 = torch.FloatTensor(data['change_30'].values)
x_3 = x1.view(50000,1)
x4 = torch.FloatTensor(data['change_40'].values)
x_4 = x4.view(50000,1)

xtrain1 = torch.stack([x_1,x_3,x_6,x_7,x_8,x_9], dim=-1)
xtrain = xtrain1.view(50000,6)
x_train = torch.FloatTensor(xtrain)
ytrain = torch.FloatTensor(TF['TF'].values)
ytrain1 = ytrain.view(50000,1)
y_train = torch.FloatTensor(ytrain1)

model = nn.Sequential(
    nn.Linear(6,30),
    nn.LeakyReLU(0.2),
    nn.Linear(30,60),
    nn.LeakyReLU(0.2),
    nn.Linear(60,60),
    nn.LeakyReLU(0.2),    
    nn.Linear(60,60),
    nn.LeakyReLU(0.2),        
    nn.Linear(60,30),
    nn.LeakyReLU(0.2),          
    nn.Linear(30,1),
)
loss_func = nn.L1Loss().to(gpu)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=0.001)
model = model.to(gpu)
x = x_train.to(gpu)
y_noise = y_train.to(gpu)
num_epoch = 180000
loss_array = []
for epoch in range(num_epoch):
    optimizer.zero_grad()
    output = model(x)
    
    loss = loss_func(output,y_noise)
    loss.backward()
    optimizer.step()
    
    loss_array.append(loss)
    if epoch % 300 == 0:
        print('epoch:', epoch, ' loss:', loss.item())
plt.plot(loss_array)
plt.show()
model_save_name = 'classifier3.pt'
path = F"/content/drive/My Drive/{model_save_name}" 
torch.save(model.state_dict(), path)
```
